```python
import pandas as pd
import numpy as np
import io
import matplotlib.pyplot as plt
import seaborn as sns

file_name = 'datos_dengue_clima_unificados.csv'

print(f"Attempting to read file '{file_name}' from the Colab environment...")

try:
    if file_name.endswith('.csv'):
        df = pd.read_csv(file_name)
    elif file_name.endswith(('.xls', '.xlsx')):
        df = pd.read_excel(file_name)
    else:
        print("Unsupported file format. Please upload a .csv or .xlsx file.")
        df = None

    if df is not None:
        print(f"\nFile '{file_name}' read successfully!")
        print(f"Original shape: {df.shape}")

        print("\n--- Data Cleaning ---")
        print("Replacing -999 values with NaN...")
        df.replace(-999, np.nan, inplace=True)
        print("Replacement complete.")

        print("Dropping rows with any missing values...")
        original_rows = df.shape[0]
        df.dropna(inplace=True)
        rows_removed = original_rows - df.shape[0]
        print(f"Dropped {rows_removed} rows containing null values.")
        print(f"New shape after cleaning: {df.shape}")

        # --- DATA RESHAPING FOR ANALYSIS ---
        print("\n--- Data Reshaping for Analysis ---")

        # 1. Isolate and reshape Dengue cases data
        print("Reshaping dengue cases data...")
        dengue_cols = ['Municipio', 'Año', 'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre']
        dengue_df = df[dengue_cols].drop_duplicates().copy()
        dengue_melted = dengue_df.melt(
            id_vars=['Municipio', 'Año'],
            var_name='Month_ES',
            value_name='Dengue_Cases'
        )

        # 2. Function to extract and reshape a specific climate parameter
        def extract_and_melt_climate(param_name, new_col_name, param_col='parameter_y', month_suffix='_y'):
            print(f"Extracting and reshaping parameter: {param_name}")
            month_cols = [m + month_suffix for m in ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']]
            id_cols = ['Municipio', 'Año']

            param_df = df[df[param_col] == param_name][id_cols + month_cols].copy()
            param_melted = param_df.melt(
                id_vars=id_cols,
                var_name='Month_EN_suffixed',
                value_name=new_col_name
            )
            param_melted['Month_EN'] = param_melted['Month_EN_suffixed'].str.replace(month_suffix, '')
            return param_melted.drop(columns=['Month_EN_suffixed'])

        # 3. Extract all required climate variables
        # Note: We assume these parameters exist in your 'parameter_y' column
        try:
            t2m_min_df = extract_and_melt_climate('T2M_MIN', 'Temp_Min')
            t2m_dew_df = extract_and_melt_climate('T2MDEW', 'Dew_Point_Temp')
            cloud_amt_df = extract_and_melt_climate('CLOUD_AMT', 'Cloud_Amount', 'parameter_x', '_x')
        except Exception as e:
            print(f"Warning: Could not extract all climate variables. Some plots might be empty. Error: {e}")
            # Create empty dataframes if extraction fails to avoid crashing
            t2m_min_df = pd.DataFrame(columns=['Municipio', 'Año', 'Month_EN', 'Temp_Min'])
            t2m_dew_df = pd.DataFrame(columns=['Municipio', 'Año', 'Month_EN', 'Dew_Point_Temp'])
            cloud_amt_df = pd.DataFrame(columns=['Municipio', 'Año', 'Month_EN', 'Cloud_Amount'])


        # 4. Merge all dataframes into a final analytical dataset
        print("Merging all datasets...")
        month_map = {
            'Enero': 'jan', 'Febrero': 'feb', 'Marzo': 'mar', 'Abril': 'apr',
            'Mayo': 'may', 'Junio': 'jun', 'Julio': 'jul', 'Agosto': 'aug',
            'Septiembre': 'sep', 'Octubre': 'oct', 'Noviembre': 'nov', 'Diciembre': 'dec'
        }
        dengue_melted['Month_EN'] = dengue_melted['Month_ES'].map(month_map)

        final_df = pd.merge(dengue_melted, t2m_min_df, on=['Municipio', 'Año', 'Month_EN'], how='left')
        final_df = pd.merge(final_df, t2m_dew_df, on=['Municipio', 'Año', 'Month_EN'], how='left')
        final_df = pd.merge(final_df, cloud_amt_df, on=['Municipio', 'Año', 'Month_EN'], how='left')
        final_df.dropna(inplace=True) # Drop rows where merge failed or data was missing
        print("Reshaping complete. Final analytical dataset is ready.")
        print(f"Final dataset shape: {final_df.shape}")

        # --- THEORIES ANALYSIS & VISUALIZATION ---
        print("\n--- Theories Analysis & Visualization ---")
        sns.set_theme(style="whitegrid")

        if not final_df.empty:
            # Theory 1: Thermal Window
            plt.figure(figsize=(12, 7))
            sns.regplot(data=final_df, x='Temp_Min', y='Dengue_Cases', scatter_kws={'alpha':0.2}, line_kws={'color':'red'})
            plt.title('Theory 1: Minimum Temperature vs. Dengue Cases')
            plt.xlabel('Minimum Temperature (T2M_MIN)')
            plt.ylabel('Dengue Cases')
            plt.show()

            # Theory 2: Absolute Humidity
            plt.figure(figsize=(12, 7))
            sns.regplot(data=final_df, x='Dew_Point_Temp', y='Dengue_Cases', scatter_kws={'alpha':0.2}, line_kws={'color':'red'})
            plt.title('Theory 2: Dew Point Temperature (Humidity) vs. Dengue Cases')
            plt.xlabel('Dew Point Temperature (T2MDEW)')
            plt.ylabel('Dengue Cases')
            plt.show()

            # Theory 3: Cloudiness and Behavior
            plt.figure(figsize=(12, 7))
            sns.regplot(data=final_df, x='Cloud_Amount', y='Dengue_Cases', scatter_kws={'alpha':0.2}, line_kws={'color':'red'})
            plt.title('Theory 3: Cloud Amount vs. Dengue Cases')
            plt.xlabel('Cloud Amount (CLOUD_AMT)')
            plt.ylabel('Dengue Cases')
            plt.show()
        else:
            print("\nFinal dataset is empty. Cannot generate plots. Please check if the parameter names (e.g., 'T2M_MIN') are correct.")


except FileNotFoundError:
    print(f"\nERROR: File '{file_name}' not found.")
    print("Please make sure the file has been uploaded to the Colab environment.")
except Exception as e:
    print(f"\nAn error occurred while processing the file: {e}")
```
