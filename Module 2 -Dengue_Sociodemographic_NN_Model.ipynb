cutting-edge technique to "ask" the model which factors were the most

```python
# ==============================================================================
# NASA Space Apps Challenge - Dengue Risk Model with Sociodemographic Data
#
# This script trains an advanced neural network to predict dengue cases based on
# sociodemographic data. It then uses the SHAP library to analyze the
# importance of each variable in the model's predictions.
# ==============================================================================

# ------------------------------------------------------------------------------
# STEP 0: ENVIRONMENT SETUP & SEED FIXING
# ------------------------------------------------------------------------------
print("--- STEP 0: INSTALLING LIBRARIES AND SETTING SEEDS ---")

# Install SHAP for model interpretability
!pip install shap -q

import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import shap
import sys
import matplotlib.pyplot as plt

# CRITICAL: Set seeds for reproducibility
seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)

print("Libraries installed and seeds set for reproducibility.")

# ------------------------------------------------------------------------------
# STEP 1: UPLOAD AND LOAD DATA
# ------------------------------------------------------------------------------
print("\n--- STEP 1: UPLOAD AND LOAD DATA ---")
print("Please select your 'dengue_sociograficos.csv' file.")

try:
    from google.colab import files
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    print(f"\nFile '{filename}' uploaded successfully.")
    df = pd.read_csv(filename)
    print("File read successfully.")
except (ValueError, IndexError, ImportError):
    print("\nUpload failed or was canceled.")
    print("Please make sure 'dengue_sociograficos.csv' is in the same directory.")
    try:
        df = pd.read_csv("dengue_sociograficos.csv")
    except FileNotFoundError:
        print("ERROR: 'dengue_sociograficos.csv' not found. Please upload the file.")
        sys.exit()

# ------------------------------------------------------------------------------
# STEP 2: DATA PREPROCESSING
# ------------------------------------------------------------------------------
print("\n--- STEP 2: PREPROCESSING DATA ---")

# Select only numeric columns for the model, but keep the target column
df_numeric = df.select_dtypes(include=[np.number])
if 'casos' in df.columns:
    df_numeric['casos'] = df['casos']
else:
    print("ERROR: Target column 'casos' not found in the original DataFrame.")
    sys.exit()

# Separate predictor variables (X) and the target variable (y)
X = df_numeric.drop(columns=['casos'])
y = df_numeric['casos']
print(f"Data separated into {X.shape[1]} features and 1 target variable.")

# --- NEW: Translate column names to English for SHAP plot ---
# Create a dictionary to map Spanish names to English
column_translation = {
    'Viviendas particulares habitadas con piso de tierra': 'Dwellings with dirt floor',
    'Viviendas particulares habitadas que disponen de letrina (pozo u hoyo)': 'Dwellings with latrine',
    'Área kilómetros cuadrados': 'Area (sq km)',
    'Viviendas particulares habitadas que no disponen de drenaje': 'Dwellings without drainage',
    'Viviendas particulares de uso temporal': 'Temporary use dwellings',
    'Población de 0 a 2 años': 'Population 0-2 years',
    'Población de 65 años y más': 'Population 65+ years',
    'Población masculina de 60 años y más': 'Male population 60+ years',
    'Viviendas particulares habitadas que disponen de cisterna o aljibe': 'Dwellings with cistern',
    'Población de 12 años y más desocupada': 'Unemployed population 12+ years',
    'Población sin afiliación a servicios de salud': 'Population without health service',
    'Población de 60 años y más': 'Population 60+ years',
    'Viviendas particulares habitadas con piso de material diferente de tierra': 'Dwellings with non-dirt floor',
    'Viviendas particulares habitadas que no disponen de agua entubada en el ámbito de la vivienda': 'Dwellings without piped water',
    'Total de hogares censales': 'Total census households',
    'Hogares censales con persona de referencia hombre': 'Households with male head',
    'Población con limitación para recordar o concentrarse': 'Population with memory/concentration limitation',
    'Viviendas particulares habitadas que disponen de drenaje': 'Dwellings with drainage',
    'Población femenina de 60 años y más': 'Female population 60+ years',
    'Población masculina': 'Male population'
}

# Rename the columns in the feature DataFrame
X.rename(columns=column_translation, inplace=True)
print("Feature names translated to English for visualization.")


# Normalize the feature data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print("Features have been scaled.")

# ------------------------------------------------------------------------------
# STEP 3: BUILD THE ADVANCED NEURAL NETWORK
# ------------------------------------------------------------------------------
print("\n--- STEP 3: BUILDING THE NEURAL NETWORK ARCHITECTURE ---")

model = Sequential([
    Dense(128, activation='relu', input_shape=(X_scaled.shape[1],)),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')  # Output layer for regression (predicting a continuous value)
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
model.summary()

# ------------------------------------------------------------------------------
# STEP 4: TRAIN THE MODEL
# ------------------------------------------------------------------------------
print("\n--- STEP 4: TRAINING THE MODEL ---")
history = model.fit(X_scaled, y, epochs=150, batch_size=16, verbose=1)
print("\nModel training complete.")

# ------------------------------------------------------------------------------
# STEP 5: CREATE A NORMALIZED RISK INDEX
# ------------------------------------------------------------------------------
print("\n--- STEP 5: GENERATING PREDICTIONS AND RISK INDEX ---")

# Generate predictions on the entire dataset
predictions = model.predict(X_scaled)

# Normalize predictions to a 0-1 scale to create a risk index
df['dengue_risk'] = (predictions - predictions.min()) / (predictions.max() - predictions.min())
print("Normalized dengue risk index created.")

# ------------------------------------------------------------------------------
# STEP 6: EXPLAIN MODEL PREDICTIONS WITH SHAP
# ------------------------------------------------------------------------------
print("\n--- STEP 6: ANALYZING FEATURE IMPORTANCE WITH SHAP ---")
print("Calculating SHAP values... (this may take a moment)")

# Use a sample of the data for the explainer for efficiency
explainer = shap.Explainer(model, X_scaled)
shap_values = explainer(X_scaled[:500]) # Using 500 samples for speed

# Generate the summary plot to visualize feature importance
print("\nDisplaying SHAP summary plot:")
plt.figure()
# Now, feature_names will be in English because we renamed the columns of X
shap.summary_plot(shap_values, features=X, feature_names=X.columns, show=False)
plt.title("Feature Importance for Dengue Prediction")
plt.xlabel("SHAP Value (impact on model output)")
plt.show()


# ------------------------------------------------------------------------------
# STEP 7: DISPLAY FINAL RESULTS
# ------------------------------------------------------------------------------
print("\n--- FINAL RESULTS PREVIEW ---")
# Show the municipality, actual cases, and the calculated risk index
print(df[['Municipio', 'casos', 'dengue_risk']].head().to_string())
print("\n--- SCRIPT COMPLETE ---")

```

Here is the technical explanation in three paragraphs, with analogies to molecular biology principles.

Data Preparation and Environmental Profiling
This initial phase of the script (Steps 0-2) is analogous to preparing a cellular sample for transcriptomic analysis. After loading essential libraries and setting random seeds for reproducibility, the code ingests the dengue_sociograficos.csv dataset, which represents the complex "environment" of each municipality. Basic data cleaning is performed by isolating numeric features, which are then separated into predictor variables (X), representing the various environmental stimuli or cellular components (e.g., protein concentrations, metabolite levels), and a target variable (y), which represents the observed phenotype or disease state (dengue cases). A critical preprocessing step is the use of StandardScaler, which normalizes each feature. This is akin to standardizing the concentrations in a biological assay, ensuring that all input signals are on a comparable scale and preventing features with large numerical ranges from disproportionately influencing the model's learning process.

The Neural Network as a Signaling Cascade
Step 3 constructs a deep neural network, which can be thought of as a model of an intracellular signaling cascade. The Sequential model is built with a series of Dense layers (128, 64, and 32 neurons), where each layer acts like a protein kinase or transcription factor that integrates multiple upstream signals to activate a downstream pathway. The relu activation function serves as the all-or-nothing activation threshold for these neurons. A key architectural feature is the inclusion of Dropout layers (with rates of 0.3 and 0.2). This technique introduces a form of stochasticity, analogous to the inherent noise in biological systems, by randomly deactivating neurons during training. This forces the network to learn robust, redundant signaling pathways, preventing it from becoming overly reliant on any single feature and thus improving its ability to generalize to new data. The final Dense layer with a linear activation consolidates all the processed signals into a single, quantitative prediction for the final phenotype—the number of dengue cases.

Model Training, Risk Indexing, and Interpretation via SHAP
The final steps (4-7) involve training the model and, crucially, interpreting its internal logic. During training (model.fit), the network is iteratively exposed to the sociodemographic data, and its internal weights are adjusted via backpropagation to minimize the error between its predictions and the actual dengue cases, effectively learning the complex regulatory logic of the system. In Step 5, the model's raw predictions are normalized to a 0-1 scale, creating an easily interpretable dengue_risk index. The most advanced part of the analysis is Step 6, which uses the SHAP (SHapley Additive exPlanations) library. This is analogous to performing a high-throughput series of in silico knockout experiments. SHAP systematically evaluates how each sociodemographic feature contributes to the model's final prediction for a given municipality. The resulting summary plot  ranks the features by their impact, making the "black box" neural network transparent and revealing which factors (e.g., 'Dwellings without drainage', 'Population without health service') are the most critical drivers of dengue risk.
